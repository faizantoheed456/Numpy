```python
def compute_loop(prices, threshold):
    result = np.where(prices > threshold, prices * 1.1, prices * 0.9)
    return result
```

```python
prices = np.array([10, 20, 30, 40, 50])
threshold = 20
```

```python
compute_loop(prices, threshold)
```
