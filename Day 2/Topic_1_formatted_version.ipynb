## 📘 Day 2, Topic 1: Boolean Masking — Filtering Data Without Loops

### What Is Boolean Masking?
A **boolean mask** is an array of `True`/`False` values that acts like a filter. When you apply it to an array, NumPy returns only the elements where the mask is `True`. This replaces the need for `if` statements inside `for` loops entirely.

### Step-by-step: How It Works
```python
temperatures = np.array([31, 22, 35, 18, 33, 27])

# Step 1: Create the mask
mask = temperatures > 30
# → [True, False, True, False, True, False]

# Step 2: Apply the mask
hot_days = temperatures[mask]
# → [31, 35, 33]    only where mask was True
```
You can also write it in one line: `temperatures[temperatures > 30]`

### Combining Conditions
Use **bitwise** operators (NOT Python's `and`/`or`/`not`!):

| Operator | Meaning | Example |
|----------|---------|---------|
| `&` | AND — both conditions true | `(a > 30) & (a < 35)` |
| `\|` | OR — at least one true | `(a < 20) \| (a >= 35)` |
| `~` | NOT — invert the mask | `~(a > 30)` |

> ⚠️ **Always wrap each condition in parentheses** when combining. `a > 30 & a < 35` is parsed as `a > (30 & a) < 35` — wrong!

### Modifying Values With a Mask
You can assign through a mask to change values in-place:
```python
temperatures[temperatures > 30] = 30   # cap all values at 30
```
This is the NumPy equivalent of: `for i in range(len(t)): if t[i] > 30: t[i] = 30`

### Counting & Aggregating With Masks
Boolean arrays are treated as **0/1 integers** by math functions:

```python
np.sum(temp > 30)    # counts how many elements satisfy the condition
np.any(temp > 30)    # True if at least one element satisfies
np.all(temp > 30)    # True if ALL elements satisfy
np.mean(temp > 30)   # proportion (fraction) that satisfy
```

### Boolean Masking on 2D Arrays
```python
grades = np.array([[85, 92], [45, 60]])

# Flat result — flattens the 2D array, returns 1D
grades[grades >= 90]   # → [92]

# Row-wise filtering — keep entire rows where condition holds for any column
grades[np.any(grades < 60, axis=1)]   # rows with at least one failing grade
grades[np.all(grades >= 60, axis=1)]  # rows where every grade passes
```
> 💡 `axis=1` means "check across columns (within each row)". `axis=0` means "check down rows (within each column)".

Practice Activity: Boolean Masking

````python
def boolean_masking(sales):
     #Extract all days where sales were above $100.
    high_sales = sales[sales > 100]
    print("1. All days where sales were greater than $100: ", high_sales)
    #Extract all days where sales were between $90 and $200 inclusive.
    moderate_sales = sales[(sales >= 90) & (sales <= 200)]
    print("2. All days where sales were in between $90 and $200 inclusive: ", moderate_sales)
    #Count how many days had sales below $100.
    low_sales_days = np.sum(sales < 100) 
    print("3. No of days where Sales goes below $100: ", low_sales_days)
    #Create a new array where any sale below $100 is increased by 20% (modify in place).
    sales[sales < 100] *= 1.2
    print("4. Sales after 20% increase for low days:", sales)
    #Find the average of sales for days that were above $150.
    avg_sales = np.mean(sales[sales > 150])
    print("5. Average of sales above $150: ", avg_sales)
````

````python
sales = np.array([120, 85, 210, 95, 300, 150, 75], dtype=np.float64)
boolean_masking(sales)
````
