Day 4, Topic 4: Masked Arrays – Handling Missing, Invalid, or Unwanted Data

```python
import numpy as np
import numpy.ma as ma
```

```python
data = np.array([1, 2, -999, 3, 4])
mask = np.array([False, False, True, False, False])
marr = ma.MaskedArray(data, mask=mask)
print(marr)
```

```python
data = np.array([1, 2, -999, 3, 4])
marr = ma.masked_where(data == -999, data)
print(marr)
```

```python
data = np.array([1.0, 2.0, np.nan, 4.0, np.inf, 5.0])
marr = ma.masked_invalid(data)
print(marr)
```

```python
data = np.array([1, 2, 4, 3, 4])
marr = ma.masked_equal(data, 4)
print(marr)
```

```python
data = np.array([1, 2, 4, 3, 4])
marr = ma.masked_less(data, 2)
print(marr)
```

```python
data = np.array([1, 2, -999, 3, 4])
marr = ma.masked_where(data == -999, data)
#keys methods that respect the mask: .mean(), .sum(), .std(), .var()
#.min(), .max(), .argmin(), .argmax()
#.prod(), .cumsum(), .cumprod()
print(marr.mean())
```

```python
print(marr.data)
print(marr.mask)
```

```python
valid_values = marr.compressed()
print(valid_values)
```

```python
valid_indices = np.where(~marr.mask)
print(valid_indices)
```

```python
filled = marr.filled(0)
print(filled)
```

```python
data_2d = np.array([[1, 2, -999],
                    [4, -999, 5],
                    [-999, 7, 9]])
marr_2d = ma.masked_equal(data_2d, -999)
print(marr_2d)
```

Task: You have a 2D array representing weekly sales data for 4 stores over 5 weeks. Missing or invalid entries are marked as -1

```python
sales_data = np.array([
    [120, 150, -1, 130, 140],   # Store 0
    [200, -1, 210, 205, -1],    # Store 1
    [-1, 300, 310, -1, 320],    # Store 2
    [400, 410, 420, 430, 440]   # Store 3
])
```

```python
#Create a masked array sales_masked where all -1 values are masked.
sales_masked = ma.masked_equal(sales_data, -1)
```

```python
#Print the masked array. Verify that -1 values appear as --.
print("Filtered data:\n", sales_masked)
```

```python
#Compute the average weekly sales for each store (row means) ignoring missing weeks. 
#Print as a list of 4 numbers.
row_mean = sales_masked.mean(axis=1)
print("Weekly sales average: ", row_mean.tolist())
```

```python
#Compute the total sales across all stores for each week (column sums) 
#ignoring missing values. Print as a list of 5 numbers.
col_sum = sales_masked.sum(axis=0)
print("Sum of all stores all week: ", col_sum.tolist())
```

```python
#Identify which stores have any missing data (i.e., their row mask contains at least one True). 
#Print their indices.
missing_indices = np.where(sales_masked.mask.any(axis=1))[0]
print("Missing data rows: ", missing_indices.tolist())
```

```python
#Fill the missing values with the mean of that store's 
#available weeks and convert back to a regular NumPy array.
filled_arrays = sales_masked.filled(row_mean[:, np.newaxis])
print(filled_arrays)
```
