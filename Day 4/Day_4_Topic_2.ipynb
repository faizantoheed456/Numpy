```python
import numpy as np
```

```python
rgb_dtype = np.dtype([
    ('r', 'u1'),
    ('g', 'u1'),
    ('b', 'u1'),
])
img_structured = np.zeros((2, 2), dtype=rgb_dtype)
print(img_structured)
```

```python
img_structured[0, 0] = (255, 0, 0)
img_structured[0, 1] = (0, 255, 0)
img_structured[1, 0] = (0, 0, 255)
img_structured[1, 1] = (255, 255, 255)
```

```python
red_channel = img_structured['r']
print(red_channel)
```

```python
# Simulate loading a 4x6 image (4 rows, 6 columns, 3 channels)
h, w = 4, 6
img_3d = np.random.randint(0, 256, size=(h, w, 3), dtype=np.uint8)
print(img_3d.shape)
```

```python
img_struct = img_3d.view(rgb_dtype).reshape(h, w)
print(img_struct.shape)
```

```python
red_mask = img_struct['r'] > 200
img_struct['r'][red_mask] = 255
img_struct['g'][red_mask] = 255
img_struct['b'][red_mask] = 255
```

```python
temp = img_struct['r'].copy()
img_struct['r'] = img_struct['b']
img_struct['b'] = temp
```

```python
#Task 1: Create a (5, 5, 3) array of random uint8 values (0–255).
imag_3d = np.random.randint(0, 256, size=(5, 5, 3), dtype=np.uint8)
```

```python
#Task 2: View it as a structured array with fields 'r', 'g', 'b'.
imag_3d_struct = imag_3d.view(rgb_dtype).reshape((5, 5))
```

```python
#Task 3: Print the shape of the structured array – it should be (5, 5).
print(f"Shape of structured array: {imag_3d_struct.shape}")
```

```python
#Task 4: Set all pixels where the red channel is > 200 to pure white (255, 255, 255). 
#Use a Boolean mask on the red channel and then assign to all three channels.
red_mask = imag_3d_struct['r'] > 200
imag_3d_struct['r'][red_mask] = 255
imag_3d_struct['g'][red_mask] = 255
imag_3d_struct['b'][red_mask] = 255
```

```python
#Task 5: Compute the average color of the image (mean of each channel separately). 
#Print as (R_avg, G_avg, B_avg).
r_avg = imag_3d_struct['r'].mean()
g_avg = imag_3d_struct['g'].mean()
b_avg = imag_3d_struct['b'].mean()
print(f"Average color: ({r_avg:.2f}, {g_avg:.2f}, {b_avg:.2f})")
```

```python
#Task 6: (Bonus) Swap the red and blue channels of the entire image using the structured view.
temp = imag_3d_struct['r'].copy()
imag_3d_struct['r'] = imag_3d_struct['b']
imag_3d_struct['b'] = temp
```
