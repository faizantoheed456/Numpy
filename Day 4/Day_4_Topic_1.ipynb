Practice Activity: Structured Arrays

```python
#TASK 1: Define a dtype with fields: 'name' (U20), 'department' (U10), 'years_exp' (i4), 'salary' (f8).
dtype = np.dtype([
    ('Name', 'U20'),
    ('Department', 'U20'),
    ('Years_exp', 'i4'),
    ('Salary', 'f8')
])
```

```python
#TASK 2: Create an array with at least 5 employees (you choose the data).
data = [
    ('Faizan', 'Data science', 4, 55000.0),
    ('Ali', 'Cyber security', 7, 75000.0),
    ('Usman', 'Artificial intelligence', 2, 180000.0),
    ('Rohan', 'Engineering', 3, 75400.0),
    ('Arslan', 'Engineering', 10, 189000.0)
]
arr = np.array(data, dtype=dtype) 
```

```python
#TASK 3: Print the names of all employees in the 'Engineering' department.
engg_names = arr[arr['Department'] == 'Engineering']
print(f"Names of employees in engineering department: {engg_names['Name']}")
```

```python
#TASK 4: Give everyone in 'Sales' a 5% raise (modify in‑place).
sales_mask = arr['Department'] == 'Sales'
arr['Salary'][sales_mask] *= 1.05
print(f"Employees salary after raising it to 5%:\n{arr[['Name', 'Salary']]}")
```

```python
#TASK 5: Sort the array by years_exp descending and print the sorted names.
sorted_sal = np.sort(arr, order='Years_exp')[::-1]
print(f"Array by year experience: {sorted_sal['Name']}")
```

```python
#Task 6: Compute the average salary per department.
departments = np.unique(arr['Department'])
for dept in departments:
    dept_mask = arr['Department'] == dept
    avg_salary = arr['Salary'][dept_mask].mean()
    print(f"{dept}: {avg_salary:.2f}")
```
