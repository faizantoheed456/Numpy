Task: Create a 1D memory‑mapped array of 10 million float64 numbers.

```python
import numpy as np
import os

# ------------------------------------------------------------
# Task 1: Create the memmap file and array
# ------------------------------------------------------------
filename = 'numbers_1.dat'
shape = (10_000_000,)
dtype = np.float64

# 'w+' creates/overwrites the file, read/write mode
arr = np.memmap(filename, dtype=dtype, mode='w+', shape=shape)

# ------------------------------------------------------------
# Task 2: Fill the array in chunks to avoid memory spikes
# ------------------------------------------------------------
chunk_size = 1_000_000
for start in range(0, shape[0], chunk_size):
    end = min(start + chunk_size, shape[0])
    arr[start:end] = np.arange(start, end, dtype=dtype)
    # Optional: print progress
    print(f"Filled indices {start} to {end-1}")

# ------------------------------------------------------------
# Task 3: Flush changes to disk and close (del releases the object)
# ------------------------------------------------------------
arr.flush()
del arr
print(f"File '{filename}' created. Size: {os.path.getsize(filename)} bytes")

# ------------------------------------------------------------
# Task 4: Reopen the file in read‑only mode
# ------------------------------------------------------------
arr_readonly = np.memmap(filename, dtype=dtype, mode='r', shape=shape)

# ------------------------------------------------------------
# Task 5: Read and print a small slice
# ------------------------------------------------------------
print("\nElements 5,000,000 to 5,000,009:")
print(arr_readonly[5_000_000:5_000_010])

# ------------------------------------------------------------
# Task 6 (Bonus): Compute total sum in chunks
# ------------------------------------------------------------
total = 0.0
chunk_size = 1_000_000
for start in range(0, shape[0], chunk_size):
    end = min(start + chunk_size, shape[0])
    total += arr_readonly[start:end].sum()
print(f"\nTotal sum of all elements: {total}")

# Expected sum: sum of 0 to 9,999,999 = 10,000,000 * 9,999,999 / 2 = 4.9999995e13
print(f"Expected sum: {10_000_000 * 9_999_999 / 2}")

# Clean up (optional)
del arr_readonly
```
