## 📘 Day 4, Topic 3: Memory-Mapped Files — Working with Arrays Too Big for RAM

### The Problem
A 10,000×10,000 float32 matrix = 400 MB. A year of sensor data at microsecond resolution = gigabytes. These don't fit in RAM. **Memory-mapped files** let you work with such arrays as if they were in memory — the OS loads only the pages you actually touch.

### How `np.memmap` Works
NumPy maps a binary file directly into the process's virtual address space. When you access a slice, the OS reads just that region from disk. When you write, changes are flushed to disk automatically (or on demand with `.flush()`).

```python
arr = np.memmap('filename.dat', dtype=np.float32, mode='w+', shape=(1_000_000,))
```

### Mode Options

| mode | Meaning | File must exist? |
|------|---------|-----------------|
| `'w+'` | Create or overwrite, read+write | No — creates it |
| `'r+'` | Open existing, read+write | Yes |
| `'r'` | Open existing, read-only | Yes |
| `'c'` | Copy-on-write (changes NOT saved) | Yes |

### Full Workflow
```python
# Step 1: CREATE — write data in chunks to avoid RAM spikes
arr = np.memmap('data.dat', dtype=np.float32, mode='w+', shape=(10_000_000,))
chunk_size = 1_000_000
for start in range(0, 10_000_000, chunk_size):
    arr[start:start+chunk_size] = np.arange(start, start+chunk_size, dtype=np.float32)
arr.flush()   # force write to disk
del arr       # release the mapping

# Step 2: REOPEN — read back later (even in a different session)
arr = np.memmap('data.dat', dtype=np.float32, mode='r', shape=(10_000_000,))
print(arr[5_000_000:5_000_010])   # only reads those 10 elements from disk
```

### Discovering Shape from File Size
If you don't know the shape of an existing file:
```python
import os
file_size = os.path.getsize('data.dat')
num_elements = file_size // np.dtype(np.float32).itemsize
arr = np.memmap('data.dat', dtype=np.float32, mode='r', shape=(num_elements,))
```

### Chunked Processing — The Key Pattern
Never load the whole file. Process in chunks:
```python
total = 0.0
chunk_size = 1_000_000
for start in range(0, shape[0], chunk_size):
    end = min(start + chunk_size, shape[0])
    total += arr[start:end].sum()   # only this chunk is in RAM
```

### Multi-Dimensional Memmaps
```python
# 2D: large matrix — fill row by row in batches
arr_2d = np.memmap('matrix.dat', dtype=np.float32, mode='w+', shape=(10000, 10000))
for i in range(0, 10000, 100):
    arr_2d[i:i+100, :] = np.random.randn(100, 10000).astype(np.float32)

# 3D: image dataset (1000 images, 256×256, RGB)
images = np.memmap('images.dat', dtype=np.uint8, mode='w+', shape=(1000, 256, 256, 3))
```

### Important Rules
- Always `del arr` (or `arr.flush()`) before reopening the file — OS needs to sync
- `mode='r'` arrays are **read-only** — assignment raises an error
- A memmap behaves like a normal NumPy array for slicing, math, broadcasting
- The file is raw binary (no header) — you must remember the `dtype` and `shape`

### When to Use Memmap vs Loading Everything
| | `np.memmap` | `np.load` / `np.fromfile` |
|--|--|--|
| Data > RAM | ✅ | ❌ |
| Random access patterns | ✅ | ❌ |
| Speed (sequential) | Similar | Similar |
| Persistent on disk | ✅ Always | Only if saved |
| Shared between processes | ✅ | Needs copy |

Task: Create a 1D memory‑mapped array of 10 million float64 numbers.

````python
import numpy as np
import os

# ------------------------------------------------------------
# Task 1: Create the memmap file and array
# ------------------------------------------------------------
filename = 'numbers_1.dat'
shape = (10_000_000,)
dtype = np.float64

# 'w+' creates/overwrites the file, read/write mode
arr = np.memmap(filename, dtype=dtype, mode='w+', shape=shape)

# ------------------------------------------------------------
# Task 2: Fill the array in chunks to avoid memory spikes
# ------------------------------------------------------------
chunk_size = 1_000_000
for start in range(0, shape[0], chunk_size):
    end = min(start + chunk_size, shape[0])
    arr[start:end] = np.arange(start, end, dtype=dtype)
    # Optional: print progress
    print(f"Filled indices {start} to {end-1}")

# ------------------------------------------------------------
# Task 3: Flush changes to disk and close (del releases the object)
# ------------------------------------------------------------
arr.flush()
del arr
print(f"File '{filename}' created. Size: {os.path.getsize(filename)} bytes")

# ------------------------------------------------------------
# Task 4: Reopen the file in read‑only mode
# ------------------------------------------------------------
arr_readonly = np.memmap(filename, dtype=dtype, mode='r', shape=shape)

# ------------------------------------------------------------
# Task 5: Read and print a small slice
# ------------------------------------------------------------
print("\nElements 5,000,000 to 5,000,009:")
print(arr_readonly[5_000_000:5_000_010])

# ------------------------------------------------------------
# Task 6 (Bonus): Compute total sum in chunks
# ------------------------------------------------------------
total = 0.0
chunk_size = 1_000_000
for start in range(0, shape[0], chunk_size):
    end = min(start + chunk_size, shape[0])
    total += arr_readonly[start:end].sum()
print(f"\nTotal sum of all elements: {total}")

# Expected sum: sum of 0 to 9,999,999 = 10,000,000 * 9,999,999 / 2 = 4.9999995e13
print(f"Expected sum: {10_000_000 * 9_999_999 / 2}")

# Clean up (optional)
del arr_readonly
````
