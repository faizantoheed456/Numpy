```python
#Problem 3.1: Normalizing Image Batch
images = np.random.rand(32, 64, 64)
means = images.mean(axis=(1,2))
result = images - means[:, np.newaxis, np.newaxis]
print(result.shape)
```

```python
#Problem 3.2: Outer Product with Broadcasting
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
result = x[:, np.newaxis] * y
print(result)
print(result.shape)
```

```python
#Problem 3.3: Add a Bias to Each Channel (Channels First)
images = np.random.rand(16, 3, 32, 32)
bias = np.array([0.1, 0.2, 0.3])
result = images + bias[np.newaxis, :, np.newaxis, np.newaxis]
print(result.shape)
```

```python
#Problem 3.4: Add a Bias to Each Channel (Channels Last)
images = np.random.rand(16, 32, 32, 3)
bias = np.array([0.1, 0.2, 0.3])
result = images + bias
print(result.shape)
```
